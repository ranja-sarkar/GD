# Gradient Descent

>> Imagine you're blindfolded on a hill, and your goal is to reach the lowest point without any knowledge of the terrain. You can achieve it by taking small steps in the direction (hill downwards) of the steepest slope (gradient).

> Steps taken to reach the 'good enough' point in the hilly landscape

1. Start at a random point: Imagine you're randomly placed.

2. Compute the slope/gradient: Determine the direction in which the hill is steepest (derivative of the function at your current location).

3. Take a step downhill: Move a small distance in the direction of the steepest slope. The size of this step is controlled by a parameter called the "learning rate."

4. Repeat: Keep repeating steps 2 and 3 until you're close enough to the lowest point.

```python
import numpy as np
import matplotlib.pyplot as plt

def cost_function(x):
    return x**2 - 4*x + 3

def gradient(x):    #derivative of the cost function
    return 2*x - 4

x = np.linspace(-10, 10, 50)
y = cost_function(x)

plt.plot(x, y, marker = '.')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Plot of cost function versus x')
plt.show()
```

```python
learning_rate = 0.1  # Step-size
iterations = 20      ##iterations


x = 0.0      #initial guess (starting point)
for i in range(iterations):
  grad = gradient(x)    #current point grad
  x = x - learning_rate * grad      #update

minimum_x = round(x, 2)
minimum_cost = round(cost_function(minimum_x), 2)
print(f"Minimum value of x: {minimum_x}")
print(f"Minimum cost: {minimum_cost}")
```

# Gradient Descent convergence

>> Cost versus iterations

```python
import numpy as np
import matplotlib.pyplot as plt

def cost_function(x):
    return x**2 - 4*x + 3

def gradient(x):           #derivative of the cost function
    return 2*x - 4

x = np.random.randn()
learning_rate = 0.1
max_iterations = 100


cost_values = []
for i in range(max_iterations):
  cost = cost_function(x)
  grad = gradient(x)
  x = x - learning_rate * grad
  cost_values.append(cost)
  if i % 10 == 0:
    print("Iteration {}: x = {}, cost = {}".format(i, x, cost))

plt.plot(cost_values)
plt.xlabel("Iterations")
plt.ylabel("Cost")
plt.show()
```
